```python
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([1, 6])
ypoints = np.array([5, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([3, 6, 9, 12])
ypoints = np.array([2, 8, 1, 10])
plt.plot(xpoints, ypoints, marker = 'o')
plt.show
```

```python
xpoints = np.array([0, 1, 2, 3, 4, 5])
ypoints = np.array([2, 4, 6, 14, 10, 12])

plt.plot(xpoints, ypoints, 'D--r', ms=10, mec='g', mfc='g')
plt.show
```

```python
y1 = np.array([1, 2, 3, 4, 5])
y2 = np.array([5, 6, 7, 8, 9])
y3 = np.array([9, 10, 11, 12, 13])

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

plt.show()
```

```python
marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}
```

```python
students = np.array(['Andy', 'Amy', 'James', 'Jules', 'Arthur'])
grades = np.array([88, 66, 90, 55, 77])

plt.bar(students, grades)
plt.title('Student Marks')
plt.show
```

```python
grades = np.array([88, 66, 90, 55, 77])

students = ['Andy', 'Amy', 'James', 'Jules', 'Arthur']
plt.title('Students Marks')
plt.pie(grades, labels=students)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show
```

```python
y1 = np.array([1, 2, 3, 4, 5])
y2 = np.array([5, 6, 7, 8, 9])
plt.subplot(2, 3, 1)
plt.plot(y1)
plt.plot(y2)

x = np.array (['A','B','C'])
y = np.array ([12, 6, 3])
plt.subplot (2, 3, 2)
plt.bar(x, y)

p = np.array([10, 30, 25, 35])
plt.subplot(2, 3, 3)
plt.pie(p)

x = np.array([3, 8, 1, 10])
y = np.array([0, 1, 2, 3])
plt.subplot(2, 3, 4)
plt.plot(x, y)
plt.grid()

x = np.array([5, 6, 9, 12])
y = np.array([52, 64, 94, 78])
plt.subplot(2, 3, 5)
plt.scatter(x, y)

x = np.random.normal(160, 20, 230)
plt.subplot(2, 3, 6)
plt.hist(x)

plt.show()
```
